# Алгоритм Брезенхема для построения отрезка

```python
import numpy as np
import matplotlib.pyplot as plt
x = 0
y = 0
x1 = 100
y1 = 100
dx = abs(x1 - x)
dy = abs(y1 - y)
s1 = np.sign(x1 - x)
s2 = np.sign(y1 - y)
changeFlag = 0;
X = []
Y = []
i = 1
```

```python
if dy > dx:
    temp = dx
    dx = dy
    dy = temp
    changeFlag = True
else:
    changeFlag = False
t = 2*dy - dx
```

```python
while i < dx:
    X.append(x)
    Y.append(y)
    while t >=0:
        if changeFlag:
            x += s1
        else:
             y += s2
        t = t - 2*dx
    if changeFlag:
        y += s2
    else:
        x += s1
    t = t + 2*dy
    i+=1
    
```

```python
plt.plot(X, Y, 'ro')
plt.show()
```

# Построение окружности

```python
r = 1000
x = 0
y = r
delta = 3 - 2*r
u = 4*x + 6
v = 10 - 4*r
X = []
Y = []
x_c = 0
y_c = 0
```

```python
while v < 10:
    X.append(x_c + x)
    Y.append(y_c - y)
    
    X.append(x_c + y)
    Y.append(y_c - x)
    
    X.append(x_c + y)
    Y.append(y_c + x)
    
    X.append(x_c + x)
    Y.append(y_c + y)
    
    X.append(x_c - x)
    Y.append(y_c + y)
    
    X.append(x_c - y)
    Y.append(y_c + x)
    
    X.append(x_c - y)
    Y.append(y_c - x)
    
    X.append(x_c - x)
    Y.append(y_c - y)
    
    if delta <= 0:
        x = x+1
        delta = delta + u
        u = u + 4
        v = v + 4
    else:
        x = x+1
        y = y-1
        delta = delta + v
        u = u + 4
        v = v + 8
    
```

```python
plt.figure(figsize=(10, 10))
plt.plot(X, Y, 'ro')
plt.show()
```
